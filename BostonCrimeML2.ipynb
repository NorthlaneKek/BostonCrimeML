# Обучаем классификатор

```python
from sklearn.neural_network import MLPClassifier
```

```python
# TRY 1 LAYER with 1 neuron
mlp_model = MLPClassifier(hidden_layer_sizes = (1,), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
```

```python
# after fitting the model we can try to check it's opportunities
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
```

```python
print(
# error on the train set
    np.mean(y_train != y_train_pred),
# error on the test set
    np.mean(y_test != y_test_pred))
```

```python
# the error on the test set is actually the same as in the train and the difference is really small.
print(np.mean(y_train != y_train_pred) - np.mean(y_test != y_test_pred))
```

```python
# TRY 1 LAYER with 2 neurons
mlp_model = MLPClassifier(hidden_layer_sizes = (2,), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
```

```python
# after fitting the model we can try to check it's opportunities
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
```

```python
print(
# error on the train set
    np.mean(y_train != y_train_pred),
# error on the test set
    np.mean(y_test != y_test_pred))
```

```python
# the error on the test set AGAIN is pretty the same. But we can see that the overall error become less due to second neuron
print(np.mean(y_train != y_train_pred) - np.mean(y_test != y_test_pred))
```

```python
# Hyper parameter manual fitting for 100 neurons on 1 layer
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
# Draw a plots to detect hyper parameneter dependence
plt.plot(alpha_arr, train_err, 'b-o', label = 'test')
plt.plot(alpha_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.min(alpha_arr), np.max(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
# alpha is L2 regularization parameter which is used to fight against overfitting. Current plot shows us that making the
# parameter less is better for the ovefitting problem. The error on the test set is less than on the train set.
# But on on the large amount of the alpha we can notice some places where the error on the test set is higher which signals
# us about overfitting.
```

```python
# A minimal error (during the hyper parameter fitting)
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
# Show an alpha which was the best from the point of "error" view
alpha_opt = alpha_arr[test_err == min_test_err]
print(alpha_opt)
```

```python
# Final training on with optimal value on 2 NEURONS
alpha_opt = alpha_opt[0]
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (100,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)

# Error rate
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

```python
# Let's try to find a dependence of model with DEFAULT alpha on the amount of NEURONS in the range
# from 1 to 100 with step 5 for 3 layers
layer_sizes = np.arange(1, 100, 5)
test_err = []
train_err = []
for size in layer_sizes:
    mlp_model = MLPClassifier(hidden_layer_sizes = (size, size, size, ), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
# Draw a plots to detect layers amount dependence
plt.plot(layer_sizes, train_err, 'b-o', label = 'test')
plt.plot(layer_sizes, test_err, 'r-o', label = 'train')
plt.xlim([np.min(layer_sizes), np.max(layer_sizes)])
plt.title('Error vs. layers')
plt.xlabel('layers')
plt.ylabel('error')
plt.legend()
pass
# we can see that increasing the amount of layers is making the model more complex and the error is less.
# but all this good stuff should be paid by the complexity and the time of the fitting.
```

Построить графики зависимости ошибки на тестовой и обучающей выборках в зависимости от количества используемых компонент. Найти наилучшее значение количества используемых главных компонент.

```python
from sklearn.decomposition import PCA
pca = PCA( n_components = 10)
pca.fit(X)
```

```python
result_pca=pca.transform(X)
result_pca.shape
```

Разбиваем данные на обучающую и тестовую выборки

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(result_pca, y, test_size = 0.33, random_state = 7)
N_test, _ = X_test.shape
N_train, _ = X_train.shape
```

Воспользуемся классификатором Extremely Random Forests с числом деревьев 100

```python
ert = ensemble.ExtraTreesClassifier(n_estimators = 100)
ert.fit(X_train, y_train)
```

```python
err_train = np.mean(y_train != ert.predict(X_train))
err_test  = np.mean(y_test  != ert.predict(X_test))

print(err_train)
print(err_test)
```

Видим, что ошибка на обучающей выборке сильно меньше, чем на тестовой, т.о. имеет место переобучение

```python
components_arr = np.arange(1, 9, 1)
test_err = []
train_err = []
oob_err = []
for components in components_arr:
    pca = PCA(n_components = components)
    X_pca_train = pca.fit_transform(X_train)
    X_pca_test = pca.transform(X_test)
    ert = ensemble.ExtraTreesClassifier(n_estimators = 100, oob_score=True, bootstrap=True)
    ert.fit(X_pca_train, y_train)
    train_err.append(np.mean(y_train != ert.predict(X_pca_train)))
    test_err.append(np.mean(y_test  != ert.predict(X_pca_test)))
    oob_err.append(1 - ert.oob_score_)
plt.plot(components_arr, train_err, 'b-o', label = 'train')
plt.plot(components_arr, test_err, 'r-o', label = 'test')
plt.plot(components_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.min(components_arr), np.max(components_arr)])
plt.title('Error vs. Components')
plt.xlabel('components')
plt.ylabel('error')
plt.legend()
pass
```
